```python
import nate
```

```python
#import CSV of tweets, indicating text column and time column

tweets = nate.import_csv('../data/IRAhandle_tweets_2.csv', text='content', time='publish_date')
```

```python
#create burst detection object

offsets = tweets.edge_burst_pipeline()
```

```python
#generate time offsets for each term-pair occurrence in the dataset

offsets.generate_offset(minimum_offsets=100)
```

```python
#use time offsets to detect bursts

bursts = offsets.burst_detection(s=2, gamma=1)
```

```python
burst_df = bursts.export_df()
```
