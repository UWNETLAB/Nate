```python
import nate
import pandas as pd
from nate.svo.graph_svo import save_svo_graph
from IPython.display import Image
```

```python
#import CSV of tweets

tweets = nate.import_csv('../data/200_tweets.csv', text='content', time='publish_date')
```

```python
#(optional) provide list of subject and object entity types to filter for

sub_tags = ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW']
obj_tags = ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW']
```

```python
#process tweets, extracting subject-verb-object triplets, and output the reults to a pandas dataframe

svo_df = tweets.svo(sub_tags, obj_tags, to_df=True)
```

```python
#preview results

svo_df.head(10)
```

```python
#create list of subject-verb-object from dataframe column

svo_list = svo_df.svo.tolist()
```

```python
#step 7: generate graph of subject-verb-object relationships in subest of data

save_svo_graph(svo_list, 'svo_net')
```
